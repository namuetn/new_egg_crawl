```python
import matplotlib.pyplot as plt


data = pd.read_csv('products.csv')
```

```python
grouped_by_brand = data.groupby('branding').size().reset_index(name='count')
grouped_by_brand
```

# CÃ¢u 1:

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_by_brand['branding'], grouped_by_brand['count'])
plt.xlabel('HÃ£ng')
plt.ylabel('Sá» lÆ°á»£ng sáº£n pháº©m')
plt.title('Sá» lÆ°á»£ng sáº£n pháº©m cá»§a má»i hÃ£ng')
plt.xticks(rotation=90)
plt.show()
```

# CÃ¢u 2: 

```python
plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=20, edgecolor='black')
plt.xlabel('GiÃ¡')
plt.ylabel('Sá» lÆ°á»£ng sáº£n pháº©m')
plt.title('PhÃ¢n bá» giÃ¡ cá»§a cÃ¡c sáº£n pháº©m')
plt.show()
```

# CÃ¢u 3: 

```python
grouped_by_brand_price = data.groupby('branding')['price'].mean().reset_index()
# Hiá»n thá» biá»u Äá» cá»t
plt.figure(figsize=(10, 6))
plt.bar(grouped_by_brand_price['branding'], grouped_by_brand_price['price'])
plt.xlabel('HÃ£ng')
plt.ylabel('GiÃ¡ trung bÃ¬nh')
plt.title('PhÃ¢n bá» giÃ¡ sáº£n pháº©m theo hÃ£ng')
plt.xticks(rotation=90)
plt.show()
```
