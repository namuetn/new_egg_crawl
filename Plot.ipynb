```python
import matplotlib.pyplot as plt


data = pd.read_csv('products.csv')
```

```python
grouped_by_brand = data.groupby('branding').size().reset_index(name='count')
grouped_by_brand
```

# Câu 1:

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_by_brand['branding'], grouped_by_brand['count'])
plt.xlabel('Hãng')
plt.ylabel('Số lượng sản phẩm')
plt.title('Số lượng sản phẩm của mỗi hãng')
plt.xticks(rotation=90)
plt.show()
```

# Câu 2: 

```python
plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=20, edgecolor='black')
plt.xlabel('Giá')
plt.ylabel('Số lượng sản phẩm')
plt.title('Phân bố giá của các sản phẩm')
plt.show()
```

# Câu 3: 

```python
grouped_by_brand_price = data.groupby('branding')['price'].mean().reset_index()
# Hiển thị biểu đồ cột
plt.figure(figsize=(10, 6))
plt.bar(grouped_by_brand_price['branding'], grouped_by_brand_price['price'])
plt.xlabel('Hãng')
plt.ylabel('Giá trung bình')
plt.title('Phân bố giá sản phẩm theo hãng')
plt.xticks(rotation=90)
plt.show()
```

# Câu 4:

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['rating'], data['price'])
plt.xlabel('Rating')
plt.ylabel('Giá')
plt.title('Mối liên hệ giữa giá sản phẩm và rating')
plt.show()
```
